This notebook compares W&B runtime durations of different model configurations.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for the two configurations
configurations = ['3 Timesteps', '9 Timesteps']
config_3_timesteps = [1.141, 2.480, 2.667, 2.413, 1.587]
config_9_timesteps = [3.304, 4.090, 3.956, 2.880, 3.252]

# Calculating means and standard deviations
means = [np.mean(config_3_timesteps), np.mean(config_9_timesteps)]
std_devs = [np.std(config_3_timesteps), np.std(config_9_timesteps)]

# Plotting the bar plot with error bars
plt.figure(figsize=(4, 3), dpi=300)
plt.bar(configurations, means, yerr=std_devs, capsize=10, color= ['blue', 'orange'])
# Adding labels and title
plt.xlabel('Experiment Setups')
plt.ylabel('Compute (runtime in hours)')
plt.title('Comparison of Compute for using 3 and 9 Timesteps')
plt.show()
```
