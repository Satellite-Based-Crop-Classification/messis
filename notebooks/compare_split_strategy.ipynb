This notebook compares 4 different split strategies for splitting the data into folds for cross-validation.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data:
# random (5-fold) band (5-fold) sechidis (5-fold) stratified (5-fold)
# 17.2% ± 1.8% 18.2% ± 2.1% 18.2% ± 1.3% 18.0% ± 1.9%

# Data for the two configurations
configurations = ['random', 'band', 'sechidis', 'stratified']
f1_tier3_means = [17.2, 18.2, 18.2, 18.0]
f1_tier3_std_devs = [1.8, 2.1, 1.3, 1.9]


# Plotting the bar plot with error bars
plt.figure(figsize=(5, 3), dpi=300)
plt.bar(configurations, f1_tier3_means, yerr=f1_tier3_std_devs, capsize=10, color=['blue', 'green', 'red', 'purple'])
# Adding labels and title
plt.xlabel('Experiment Setups')
plt.ylabel('F1 Score (%)')
plt.title('F1 Score on Tier 3 for different fold splitting strategies')
plt.show()
```
